```python
import numpy as np
import pandas as pd
noaaNZ = pd.read_csv('data/NOAAdata/NzNoaa_3485743.csv', parse_dates=['DATE'])
rain = pd.read_csv('data/statsNZ/rainfall-data-to-2022-statsnz/daily-rainfall-at-30-sites-1960-to-2022.csv', parse_dates=['date'])
temperature1 = pd.read_csv('data/statsNZ/temperature-data-to-2022-statsnz/daily-temperature-for-30-sites-to-2022-part1.csv', parse_dates=['date'])
temperature2 = pd.read_csv('data/statsNZ/temperature-data-to-2022-statsnz/daily-temperature-for-30-sites-to-2022-part2.csv', parse_dates=['date'])
temperature3 = pd.read_csv('data/statsNZ/temperature-data-to-2022-statsnz/daily-temperature-for-30-sites-to-2022-part3.csv', parse_dates=['date'])
noaaNZ.info()
```

```python
aucklandNoaa = noaaNZ[noaaNZ['NAME'] == 'AUCKLAND AERO AWS, NZ']
aucklandNoaa.rename(columns={"DATE": 'date', "PRCP" : 'precipitation', "TAVG" : 'tavg', 'TMAX': 'tmax', 'TMIN': 'tmin' }, inplace=True)
aucklandNoaa
```

```python
filepath = 'data/aucklandNoaa.csv'
aucklandNoaa.to_csv(filepath, index=False)
```

```python
aucklandTemp = temperature1[temperature1['site'] == 'Auckland (Auckland)']
aucklandTemp
```

# Seperate the statistic column, convert from multiple rows to multiple columns

```python
pivoted_df = aucklandTemp.pivot(index=['date'], columns='statistic', values='temperature')
pivoted_df
```

```python
aucklandRain = rain[rain['site'] == 'Auckland (Auckland)']
aucklandRain
```

# Merge Rainfall data, and Temperature data

```python
auckland = aucklandRain.merge(pivoted_df, on='date')
auckland.head(10)
```

```python
auckland.rename(columns={"rainfall": 'rainfall(mm)',  "Average" : 'tavg', 'Maximum': 'tmax', 'Minimum': 'tmin' }, inplace=True)
auckland
```

```python
auckland.drop(columns=['site','rainfall_units', 'lat', 'lon'], inplace=True)
auckland
```

```python
filepath = 'data/auckland_rain_temp_statsNZ.csv'
auckland.to_csv(filepath, index=False)
```
