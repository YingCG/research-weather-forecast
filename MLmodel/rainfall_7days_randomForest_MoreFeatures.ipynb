```python
# Create lag features for prevous 7 days of rainfall and temperature
lag_days = 3

for i in range(1, lag_days - 1):
    auckland_rain[f'daybefore_rainfall{i}'] = auckland_rain['Amount(mm)'].shift(i)
    auckland_rain[f'daybefore_period{i}'] = auckland_rain['Period(min)'].shift(i)
    auckland_rain[f'daybefore_Deficit(mm){i}'] = auckland_rain['Deficit(mm)'].shift(i)
    auckland_rain[f'daybefore_Tmax(C){i}'] = auckland_rain['Tmax(C)'].shift(i)
    auckland_rain[f'daybefore_Tmin(C){i}'] = auckland_rain['Tmin(C)'].shift(i)
    auckland_rain[f'daybefore_Tgmin(C){i}'] = auckland_rain['Tgmin(C)'].shift(i)
    auckland_rain[f'daybefore_Twet(C){i}'] = auckland_rain['Twet(C)'].shift(i)
    auckland_rain[f'daybefore_Tdew(C){i}'] = auckland_rain['Tdew(C)'].shift(i)
    
for i in range(1, lag_days + 1):
    auckland_rain[f'nextday_rainfall{i}'] = auckland_rain['Amount(mm)'].shift(i)
    auckland_rain[f'nextday_period{i}'] = auckland_rain['Period(min)'].shift(i)
    auckland_rain[f'next_Deficit(mm){i}'] = auckland_rain['Deficit(mm)'].shift(i)
    auckland_rain[f'next_Tmax(C){i}'] = auckland_rain['Tmax(C)'].shift(i)
    auckland_rain[f'next_Tmin(C){i}'] = auckland_rain['Tmin(C)'].shift(i)
    auckland_rain[f'next_Tgmin(C){i}'] = auckland_rain['Tgmin(C)'].shift(i)
    auckland_rain[f'next_Twet(C){i}'] = auckland_rain['Twet(C)'].shift(i)
    auckland_rain[f'next_Tdew(C){i}'] = auckland_rain['Tdew(C)'].shift(i)

# Drop rows with NaN values due to lag features
auckland_rain.dropna(inplace=True)

auckland_rain.head(10)
```

```python
from sklearn.model_selection import train_test_split

# Split the data into features and target variable
# Features (X) and target variable (y)
features = ['Pmsl(hPa)', 'Pstn(hPa)', 'Amount(mm)', 'Period(min)',
       'daily_total_rainfall', 'daily_avg_duration', 'daily_min_duration',
       'daily_max_duration', 'daily_median_duration', 'Rainfall(mm)',
       'Deficit(mm)', 'Sunshine(Hrs)', 'WindDir(DegT)', 'Speed(m/s)_x',
       'WindDir StdDev', 'WindSpd StdDev', 'Tmax(C)', 'Tmin(C)', 'Tgmin(C)',
       'Tmean(C)', 'Twet(C)', 'RH(%)', 'Tdew(C)', 'Dir(DegT)', 'Speed(m/s)_y',
       'Percent(%)', 'soil_avg_moist', 'soil_min_moist', 'soil_max_moist',
       'soil_median_moist', 'daybefore_rainfall1', 'daybefore_period1',
       'daybefore_Deficit(mm)1', 'daybefore_Tmax(C)1', 'daybefore_Tmin(C)1',
       'daybefore_Tgmin(C)1', 'daybefore_Twet(C)1', 'daybefore_Tdew(C)1', 'nextday_period1', 'next_Deficit(mm)1',
       'next_Tmax(C)1', 'next_Tmin(C)1', 'next_Tgmin(C)1', 'next_Twet(C)1',
       'next_Tdew(C)1', 'nextday_rainfall2', 'nextday_period2',
       'next_Deficit(mm)2', 'next_Tmax(C)2', 'next_Tmin(C)2', 'next_Tgmin(C)2',
       'next_Twet(C)2', 'next_Tdew(C)2', 'nextday_rainfall3',
       'nextday_period3', 'next_Deficit(mm)3', 'next_Tmax(C)3',
       'next_Tmin(C)3', 'next_Tgmin(C)3', 'next_Twet(C)3', 'next_Tdew(C)3',]

X = auckland_rain[features]

# Define the target variable (next day's rainfall)
y = auckland_rain['nextday_rainfall1']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

```

# Random Forest

```python
# Random Forest model
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt

rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
rf_rmse = sqrt(mean_squared_error(y_test, rf_predictions))
print(f'Random Forest RMSE: {rf_rmse}')
```

```python
# Predict rainfall for tomorrow using the Random Forest model
rf_prediction_tomorrow = rf_model.predict(X_train)
rf_prediction_day_after_tomorrow = rf_model.predict(X_train)
rf_prediction_two_days_after_tomorrow = rf_model.predict(X_train)

# Print the predictions
print(f'Random Forest Prediction for tomorrow: {rf_prediction_tomorrow[0]} mm')
print(f'Random Forest Prediction for the day after tomorrow: {rf_prediction_day_after_tomorrow[0]} mm')
print(f'Random Forest Prediction for two days after tomorrow: {rf_prediction_two_days_after_tomorrow[0]} mm')

```

```python
# Calculate monthly median rainfall
monthly_median_rainfall = auckland_rain.groupby(auckland_rain['Date(NZST)'].dt.month)['Amount(mm)'].median()

# Calculate the percentage of actual rainfall compared to the monthly median
auckland_rain['rainfall_percentage'] = (auckland_rain['Amount(mm)'] / monthly_median_rainfall[auckland_rain['Date(NZST)'].dt.month].values) * 100

# Create a function to categorize the weather conditions
def categorize_rainfall_condition(percentage):
    if percentage > 200:
        return "Very wet"
    elif percentage > 110:
        return "Moderately wet"
    elif percentage > 90:
        return "Near normal"
    elif percentage > 50:
        return "Moderately dry"
    else:
        return "Very dry"

# Apply the categorization function to create a new column with the weather condition
auckland_rain['weather_condition'] = auckland_rain['rainfall_percentage'].apply(categorize_rainfall_condition)

# Print the resulting dataframe with weather conditions
print(auckland_rain[['Date(NZST)', 'Amount(mm)', 'rainfall_percentage', 'weather_condition']])
```
