```python
import pandas as pd
noaaData = pd.read_csv('../../data/statsNZ/noaa/NOAA.csv')
noaaData.head()
```

```python
noaaData = noaaData[noaaData['NAME'] == 'AUCKLAND AERO AWS, NZ']
noaaData.head()
```

```python
noaaData.info()
```

```python
noaaData['DATE'] = pd.to_datetime(noaaData['DATE'], format='%d/%m/%Y')

# Filter the data for the year 2023
noaaData_2023 = noaaData[noaaData['DATE'].dt.year == 2023]
noaaData_2023.head()
```

```python
# Function to convert Fahrenheit to Celsius
fahrenheit_to_celsius = lambda f: round((f - 32) / 1.8, 2)

# Applying the lambda function to the 'temperature_fahrenheit' column
noaaData_2023[['TAVG', 'TMAX', 'TMIN']] = noaaData_2023[['TAVG', 'TMAX', 'TMIN']].apply(fahrenheit_to_celsius)
print(noaaData_2023)
```

```python
noaaData_2023.info()
```

```python
# noaaData_2023 = noaaData_2023.drop(columns=['STATION', 'NAME', 'SNWD'])
noaaData_2023.head()
```

```python
# Conversion factor
# inches_to_mm = 25.4
# Function to convert inches to millimeters
inches_to_mm = lambda inches: round(inches * 25.4, 2)

# Applying the lambda function to the 'PRCP' column
noaaData_2023['PRCP'] = noaaData_2023['PRCP'].apply(inches_to_mm)

noaaData_2023.head()
```
