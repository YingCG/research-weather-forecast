# Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_niwa = pd.read_csv("../AucklandWeatherFromNiwa1971-2023.csv")
df_niwa.info()
df_niwa.columns
```

```python
aucklandWeather = df_niwa.rename(columns={'Tmax ©':'Tmax (C)'})
aucklandWeather.head()
```

```python
aucklandWeather.info()
```

```python
numeric_columns = ['Wdir (Deg)', 'WSpd (m/s)', 'GustDir (Deg)', 'GustSpd (m/s)',
       'Tdry (C)', 'Twet (C)', 'RH (%)', 'Tmax (C)', 'Tmin (C)', 'Tgmin (C)',
       'ET10 (C)', 'Pmsl (hPa)', 'Rad (MJ/m2)', 'Rain (mm)']

for col in numeric_columns:
    aucklandWeather[col] = pd.to_numeric(aucklandWeather[col], errors='coerce')

# Check the DataFrame information after conversion
aucklandWeather.info()
```

```python
aucklandWeather.isnull().sum()
```

```python
aucklandWeather.head()
```

```python
aucklandWeather['Date'] = pd.to_datetime(aucklandWeather['Date'], format="%d/%m/%Y")
aucklandWeather.info()
```

```python
# Heatmap for correlation between variables
import seaborn as sns

weatherAuckland = aucklandWeather.set_index('Date')
# Select only the numeric columns from the 'weather' DataFrame
numeric_columns = weatherAuckland.select_dtypes(include=[np.number])

plt.figure(figsize=(10, 8))
sns.heatmap(weatherAuckland.corr(), annot=True)
plt.show()
```

```python
# Wdir (Deg), WSpd (m/s), Tdry (C), Pmsl (hPa):
# The number of missing values for these parameters is relatively small. Using mean imputation is a simple and common approach when dealing with small amounts of missing data.

aucklandWeather['Wdir (Deg)'].fillna(aucklandWeather['Wdir (Deg)'].mean(), inplace=True)
aucklandWeather['WSpd (m/s)'].fillna(aucklandWeather['WSpd (m/s)'].mean(), inplace=True)
aucklandWeather['Tdry (C)'].fillna(aucklandWeather['Tdry (C)'].mean(), inplace=True)
aucklandWeather['Pmsl (hPa)'].fillna(aucklandWeather['Pmsl (hPa)'].mean(), inplace=True)
```

```python
# GustDir (Deg), GustSpd (m/s), Tmax (C), Tmin (C), ET10 (C):
# Linear interpolation is a suitable method when dealing with time series or sequential data.

aucklandWeather['GustDir (Deg)'].interpolate(method='linear', inplace=True)
aucklandWeather['GustSpd (m/s)'].interpolate(method='linear', inplace=True)
aucklandWeather['Tmax (C)'].interpolate(method='linear', inplace=True)
aucklandWeather['Tmin (C)'].interpolate(method='linear', inplace=True)
aucklandWeather['ET10 (C)'].interpolate(method='linear', inplace=True)
```

```python
# Twet (C), RH (%), Tgmin (C), Rad (MJ/m2), Rain (mm):
# The number of missing values is relatively high. These parameters are likely influenced by various factors,  Linear interpolation is used as a first step, but considering the higher degree of missingness, more advanced techniques or model-based imputation may be necessary. 

aucklandWeather['Twet (C)'].interpolate(method='linear', inplace=True)
aucklandWeather['RH (%)'].interpolate(method='linear', inplace=True)
aucklandWeather['Tgmin (C)'].interpolate(method='linear', inplace=True)
aucklandWeather['Rad (MJ/m2)'].interpolate(method='linear', inplace=True)
aucklandWeather['Rain (mm)'].interpolate(method='linear', inplace=True)
```
