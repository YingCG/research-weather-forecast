# Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_niwa = pd.read_csv("../AucklandWeatherFromNiwa1971-2023.csv")
df_niwa.info()
df_niwa.columns
```

```python
aucklandWeather = df_niwa.rename(columns={'Tmax ©':'Tmax (C)'})
aucklandWeather.head()
```

```python
aucklandWeather.info()
```

```python
aucklandWeather['Date'] = pd.to_datetime(aucklandWeather['Date'], format='')
```

```python
numeric_columns = ['Wdir (Deg)', 'WSpd (m/s)', 'GustDir (Deg)', 'GustSpd (m/s)',
       'Tdry (C)', 'Twet (C)', 'RH (%)', 'Tmax (C)', 'Tmin (C)', 'Tgmin (C)',
       'ET10 (C)', 'Pmsl (hPa)', 'Rad (MJ/m2)', 'Rain (mm)']

for col in numeric_columns:
    aucklandWeather[col] = pd.to_numeric(aucklandWeather[col], errors='coerce')

# Check the DataFrame information after conversion
aucklandWeather.info()
```

```python
aucklandWeather.isnull().sum()
```

```python
aucklandWeather.head()
```

```python
# Heatmap for correlation between variables
import seaborn as sns

weatherAuckland = aucklandWeather.set_index('Date')
# Select only the numeric columns from the 'weather' DataFrame
numeric_columns = weatherAuckland.select_dtypes(include=[np.number])

plt.figure(figsize=(20, 10))
sns.heatmap(weatherAuckland.corr(), annot=True)
plt.show()
```
