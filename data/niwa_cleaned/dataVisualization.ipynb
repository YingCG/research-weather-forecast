```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
weatherAuckland = pd.read_csv("./niwaDailyWeatherData.csv", parse_dates=['Date(NZST)'])
weatherAuckland.info()
```

```python
weatherAuckland.drop(columns=['Depth(cm)', 'Unnamed: 0'], inplace=True)
```

```python
weatherAuckland.to_csv('aucklandDailyWeatherData.csv')
```

```python
weatherAuckland.plot(figsize=(20, 10))
```

```python
ax = weatherAuckland.plot(figsize=(20, 8), ls='--', lw=1)
ax.legend(loc=0, bbox_to_anchor=(1,1))
```

```python
# Heatmap for correlation between variables
import seaborn as sns

weatherAuckland = weatherAuckland.set_index('Date(NZST)')
# Select only the numeric columns from the 'weather' DataFrame
numeric_columns = weatherAuckland.select_dtypes(include=[np.number])

plt.figure(figsize=(20, 10))
sns.heatmap(weatherAuckland.corr(), annot=True)
plt.show()
```
