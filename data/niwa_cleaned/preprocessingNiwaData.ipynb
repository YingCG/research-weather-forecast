```python
rainBreakpoint['Date(NZST)'] = pd.to_datetime(rainBreakpoint['Date(NZST)_Incl_secs'], format='%Y%m%d:%H%M%S').dt.date
rainBreakpoint = rainBreakpoint[['Date(NZST)', 'Amount(mm)', 'Period(min)']]
print(rainBreakpoint)
```

```python
rainBreakpoint['Amount(mm)'] = pd.to_numeric(rainBreakpoint['Amount(mm)'], errors='coerce')
rainBreakpoint['Period(min)'] = pd.to_numeric(rainBreakpoint['Period(min)'], errors='coerce')
rainBreakpoint.info()
```

```python
# Aggregation by Sum:
rainBreakpoint['daily_total_rainfall']= rainBreakpoint.groupby('Date(NZST)')['Amount(mm)'].transform('sum')

# Aggregation by Mean (Average):
rainBreakpoint['daily_avg_duration'] = rainBreakpoint.groupby('Date(NZST)')['Period(min)'].transform('mean')

# Aggregation by Min (Minimum):
rainBreakpoint['daily_min_duration'] = rainBreakpoint.groupby('Date(NZST)')['Period(min)'].transform('min')

# Aggregation by Max (Maximum)
rainBreakpoint['daily_max_duration'] = rainBreakpoint.groupby('Date(NZST)')['Period(min)'].transform('max')

# Aggregation by Median: The median is the middle value in a sorted set of values. 
rainBreakpoint['daily_median_duration'] = rainBreakpoint.groupby('Date(NZST)')['Period(min)'].transform('median')

# Drop duplicate rows created by the transform
rainBreakpoint = rainBreakpoint.drop_duplicates(subset=['Date(NZST)'])
rainBreakpoint
```

```python
rainBreakpoint.isnull().sum()
```

```python
rainBreakpoint = rainBreakpoint.dropna(subset=['Date(NZST)', 'daily_total_rainfall', 'daily_avg_duration', 'daily_min_duration', 'daily_max_duration', 'daily_median_duration'])
rainBreakpoint.isnull().sum()
```
